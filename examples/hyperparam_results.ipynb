## Hyperparameter Tuning on Look-back Horizon n

            The folowing parameters are fixed:
            hr = 'KF'
            constant = True
            metric = 'sharpe_ratio'
            signal, entry, exit = 'ecdf', 0.05, 0.5
            window = 30
            select best 10 stocks.
            The following params are tuned:
            n: look back years
            n_largest: n largest stocks in sharpe ratio to select
            n_sample" number of stocks
            unique_tickers: require unique stocks acrossing multple pairs or not

```python
res1 =pd.read_pickle('res1.pkl')
res1[['n', 'n_sample', 'rand_state', 'unique_tickers']] = pd.DataFrame(res1['params'].tolist())
res1 = res1.drop('params', axis = 1)
res1.head()
```

```python
# get 1.
res1.groupby(['n_sample', 'n', 'unique_tickers', 'n_largest']).mean().reset_index().drop('rand_state', axis = 1)
# get 2.
res1.groupby(['n_sample', 'n', 'n_largest', 'unique_tickers', ]).mean().reset_index().drop('rand_state', axis = 1)
# get 3
res1.loc[res1['unique_tickers'] == True].groupby(['n_sample', 'n', 'n_largest']).mean()\
.reset_index().drop(['rand_state', 'unique_tickers'], axis = 1)
# get 4
res1.loc[res1['unique_tickers'] == True].groupby(['n_sample', 'n', 'n_largest']).min()\
.reset_index().drop(['rand_state', 'unique_tickers', 'pairs'], axis = 1)
res1.loc[res1['unique_tickers'] == True].groupby(['n_sample', 'n', 'n_largest']).max()\
.reset_index().drop(['rand_state', 'unique_tickers', 'pairs'], axis = 1)
```

1. largest n_largest always increases performance. Thus, I might need to further increase n_largest to get better perfermance.
2. unique_tickers=false can generate large negative cumulative returns. The reseason is, with unique_tickers=false, one stock can be selected across multipple pairs, and then one wrong bet can wipe out portfolio. In a word, poor diversification with unique_tickers=Flase. Thus, the recommendation is to use unique_tickers = True always.
3. a. When performance is measured by sharpe ratio, there is a matching principle: small(large) n_sample matches with small(large) look-balck period generate better result. This might be because of overfitting when n_sample and look-back period are mismatched.
3. b. When performance is measured by cumulative return, small n_sample and small n have higher cum_r.
4. small-small have higher max_cumr depending on sample, yet min_cum_r is roughly the same as large-large.

It  seems that small look-back period with large cap is the way to go.
